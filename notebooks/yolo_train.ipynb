```python
import torch
from PIL import Image
from ultralytics import YOLO
import cv2

model = YOLO("/content/yolov8m.pt")
```

```python
model.train(data="/content/train_cfg.yaml",
            model="/content/yolov8m.pt",
            patience=5,
            batch=16,
            save=True,
            verbose=True,
            project="hack_yolov8m",
            name="platform_detection"
            )
```

```python
from ultralytics.utils.plotting import Annotator

results = model("/content/dataset_yolo/obj_val_data/image530.jpg")

for r in results:

    annotator = Annotator(cv2.imread("/content/dataset_yolo/obj_val_data/image530.jpg"))

    boxes = r.boxes
    for box in boxes:
        print(box)
        b = box.xyxy[0]  # get box coordinates in (top, left, bottom, right) format
        c = box.cls
        annotator.box_label(b, model.names[int(c)])

img = annotator.result()
cv2.imshow(img)
```
